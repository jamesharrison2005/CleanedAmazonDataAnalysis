##Sales Performance Analysis
Grouping features together can help provide insight on the performance of certain products which can allow us to find answers to important questions such as:
- Which products have the highest amount of positive ratings?
- which products have the highest estimated revenue?
- Which Brands bring have the highest ratings?

### Products with the Most Positive Ratings
To determine which products have the highest amount of positive ratings, we filter the DataFrame for 'Positive' sentiment and then sum the 'no_of_ratings' for each product.

```python
positive_ratings_by_product = df[df['sentiment'] == 'Positive'].groupby('name')['no_of_ratings'].sum().sort_values(ascending=False)
print("Products with the most positive ratings:\n", positive_ratings_by_product.head(10))
```

###Products with the highest estimated revenue


```python
highest_revenue_products = df.groupby('name')['estimated_revenue'].sum().sort_values(ascending=False)
display(highest_revenue_products)
```

###Highest estimated revenue with formatting

```python
formatted_highest_er = df.groupby('name')['estimated_revenue'].sum().apply(lambda x: f"{x / 1_000_000_000:.2f} billion").sort_values(ascending=False)
display(formatted_highest_er.head(10))
```
